**Task 1: Implement Kafka Producers**

```python
from kafka import KafkaProducer
import json
import pandas as pd
```

```python
# Sample data
data = pd.read_json("data.json")

# Kafka producer for inventory orders
inventory_producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                   value_serializer=lambda x: json.dumps(x).encode('utf-8'))

# Kafka producer for delivery orders
delivery_producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: json.dumps(x).encode('utf-8'))

## It should filter and send  messages where type is inventory to inventory_producer 
for item in data:
    if item['type'] == 'inventory':
        inventory_producer.send(f'Message delivered to inventory_orders [{item}]')

inventory_producer.flush()

```

```python
# It should filter and send  messages where type id delivery to delivery_producer
for item in data:   
    if item['type'] == 'delivery':
        delivery_producer.send('Message delivered to delivery_orders [{item}]')

delivery_producer.flush()
```

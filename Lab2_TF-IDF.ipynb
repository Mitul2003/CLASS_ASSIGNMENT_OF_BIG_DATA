```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Text-preprocessing
df1 = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))

# vectorize data
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X = vectorizer.fit_transform(df1.data)
X.shape
```

## Q-2
Create a function to calculate the cosine similarity between two TF-IDF vectors.

```python
def cosine(vec1, vec2):

    similarity_matrix = cosine_similarity(vec1, vec2)
    similarity_score = similarity_matrix[0, 0]  # Extract the similarity score from the matrix
    return similarity_score

similarity_score = cosine(X, X)  # we use same sparce matrix so similarity between two vector must be one
print("Cosine Similarity:", similarity_score)
```

## Q-3
Implement a document similarity search function that takes a document as input
and returns a list of documents ranked by their similarity to the input document.

```python
def document_similarity_search(target_vector, tfidf_matrix, dataset):

    similarity_scores = cosine_similarity(target_vector, tfidf_matrix)

    ranked_documents = [(doc, score) for doc, score in zip(dataset.data, similarity_scores[0])]
    ranked_documents.sort(key=lambda x: x[1], reverse=True)

    return ranked_documents
```

```python
# target_document = "Your input document goes here."

# also we can use inbuilt data
target_document = df1.data[0]  # You can change the index as needed

# Vectorize the target document
target_vector = vectorizer.transform([target_document])
ranked_documents = document_similarity_search(target_vector, X, df1)

# Print the top N documents and their similarity scores
top_n = 5
for i, (doc, score) in enumerate(ranked_documents[:top_n], 1):
    print(f"Rank {i}: Cosine Similarity {score:.4f}\n{doc}\n{'*'*50} {'+'*50} {'-'*50}")
```

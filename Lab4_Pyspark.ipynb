## TASK-1: Generate 100 random numbers and calculate their frequency

```python
#create spark context
sc = SparkContext("local", "RDD Tasks-1")
```

```python
# Generate 100 random numbers in range 0 to 10 using numpy randint function with the seed set to 10.
random_numbers = np.random.RandomState(seed=10).randint(0, 11, size=100)
```

```python
# Create an RDD using the parallelize function with the generated data
rdd_numbers = sc.parallelize(random_numbers)
```

```python
# Calculate the frequency of each number (0 - 10)
frequency_counts = rdd_numbers.countByValue()

# Display the results
for num, count in sorted(frequency_counts.items()):
    print(f"Number {num}: {count} occurrences")
```

## TASK-2: Calculate the frequency of each word in text8 dataset

```python
# Initialize SparkSession
spark = SparkSession.builder.appName("Task-2").getOrCreate()
```

```python
spark
```

```python
# Calculate the frequency of each word in text8 dataset using DataFrame
text_rdd = spark.sparkContext.textFile("/content/drive/MyDrive/Colab Notebooks/SEM2/BIG DATA/Lab4_data/text8")
```

```python
# Split each line into words and flatten the structure
words_rdd = text_rdd.flatMap(lambda line: line.split())
```

```python
# Calculate the frequency of each word
word_frequencies = words_rdd.countByValue()
```

```python
# Filter the RDD to get frequencies of words containing the letter 'a'
filtered_word_frequencies = words_rdd.filter(lambda word: 'a' in word).countByValue()
```

```python
# Display the overall word frequencies
for word, count in sorted(word_frequencies.items()):
    print(f"Word '{word}': {count} occurrences")
```

```python
# Display the word frequencies containing the letter 'a'
for word, count in sorted(filtered_word_frequencies.items()):
    print(f"Word '{word}': {count} occurrences")
```
